```python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
from scikitplot.cluster import plot_elbow_curve
```

```python
data=pd.read_excel("BA_New.xlsx")
```

```python
data.head()
```

```python
model_cluster = KMeans(5,random_state=42)
```

```python
model_cluster.fit_predict(data)
```

```python
model_cluster.cluster_centers_
```

```python
model_cluster.inertia_
```

```python
sse={}
for k in range(1,11):
    cluster=KMeans(k,random_state=42)
    cluster.fit_predict(data)
    sse[k] = cluster.inertia_
```

```python
print(sse)
```

```python
plt.figure()
plt.plot(sse.keys(),sse.values(), marker="o")
plt.xlabel("Number of cluster")
plt.ylabel("SSE")
plt.show()
```

```python
cluster_final=KMeans(4,random_state=42)
data["cluster_final"]=cluster_final.fit_predict(data)
```
